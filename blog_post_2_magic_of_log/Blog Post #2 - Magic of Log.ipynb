### Import Libraries

```python
# Import libraries
import numpy as np
import pandas as pd
import ipywidgets as widgets
from IPython import display
import matplotlib.pyplot as plt

%matplotlib inline
```

### Construct Dummy Datasets for Demonstration

```python
# Construct data for the demostration
count = np.array([100,500,800,1000,800,500,100])
num = np.array([0.001, 0.01, 0.1, 1, 10, 100, 1000])
log = np.log(num).round(1)
sqrt = np.sqrt(num).round(2)
```

```python
# Construct three datasets
df_1 = pd.DataFrame({'num': num, 'count': count})
df_2 = pd.DataFrame({'log': log, 'count': count})
df_3 = pd.DataFrame({'sqrt': sqrt, 'count': count})
```

```python
# create output widgets
widget1 = widgets.Output()
widget2 = widgets.Output()
widget3 = widgets.Output()

# render in output widgets
with widget1:
    display.display(df_1)
with widget2:
    display.display(df_2)
with widget3:
    display.display(df_3)

# create HBox
hbox = widgets.HBox([widget1, widget2, widget3])

# render hbox
hbox
```

### Plot the Medians

```python
# Construct the lists for histagrams
num_list = []
for i in range(len(num)):
    num_list += [num[i]]*count[i]

sqrt_list = []
for i in range(len(sqrt)):
    sqrt_list += [sqrt[i]]*count[i]
    
log_list = []
for i in range(len(log)):
    log_list += [log[i]]*count[i]
    
dataset_list = [num_list, sqrt_list, log_list]
title_list = ['Number', 'Square Root of Number', 'Log of Number']
```

```python
# Function for plotting the medians
'''Input: 
        dataset_list: a list of list (numbers)
        title_list: a list of strings
   output: plot of median as a red verticle line
           for each list of numbers
   '''
def plot_median(dataset_list, title_list):
    plt.figure(figsize = (18, 4))
    for i in range(3):
        plt.subplot(1, 3, i+1)
        plt.axvline(np.median(data_list[i]), color = 'r')
        plt.xlim(min(dataset_list[i])-1, max(dataset_list[i])+1)
        plt.title(f'Median: {title_list[i]}', size = 15)
        plt.xlabel(f'Dataset {i + 1}', size = 15)
        plt.yticks([])
    plt.tight_layout()
```

```python
# Run the function
plot_median(dataset_list, title_list)
```

### Plot the Histagram

```python
# Function to plot the histagram
'''Input: 
        dataset_list: a list of list (numbers)
        title_list: a list of strings
   output: histagram plots
   '''
def plot_hist(dataset_list, title_list):
    plt.figure(figsize = (18, 5))
    for i in range(3):
        plt.subplot(1, 3, i+1)
        plt.hist(data_list[i], bins=20)
        plt.title(f'{title_list[i]} Distribution', size=15)
        plt.xlabel(f'Dataset {i+1}', size=15)
        plt.ylabel('Count')
    plt.tight_layout()
```

```python
# Plot the histagram
plot_hist(dataset_list, title_list)
```
